```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Defining the file path
file_path = '../data/Online_Retail.xlsx'

# Loading into pandas DataFrame
df = pd.read_excel(file_path)

df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df['Description'].fillna('Unknown', inplace=True)
df['CustomerID'].fillna(0, inplace=True)
```

```python
# convert customer id to integer
df['CustomerID'] = df['CustomerID'].astype(int)
```

```python
df[['Quantity', 'UnitPrice']].describe()
```

```python
df.nunique()
```

```python
duplicates = df.duplicated().sum()
print(f"Total number of duplicates is {duplicates}")
```

```python
df = df.drop_duplicates()
```

```python
print(len(df))
```

```python
negative_qty = df[df['Quantity'] < 0].shape[0]
print("Total number of return quantity:", negative_qty)
```

```python
# creating revenue column
df['Revenue'] = df['Quantity'] * df['UnitPrice']
```

```python
df.columns
```

```python
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['MonthName'] = df['InvoiceDate'].dt.month_name()
df['Quarter'] = df['InvoiceDate'].dt.quarter
df['Date'] = df['InvoiceDate'].dt.date
```

```python
df['Weekday'] = df['InvoiceDate'].dt.day_name()
df.head()
```

```python
df[df['Revenue'] > 0].head()
```

```python
df.to_csv("../data/online_retail_cleaned.csv", index=False)
```
